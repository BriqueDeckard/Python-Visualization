```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
print("setup complete")
```

```python
df = pd.read_csv("final_depression_dataset_1.csv")
df["suicidal_thoughts"] = df["Have you ever had suicidal thoughts ?"]
df.drop(columns=["Have you ever had suicidal thoughts ?"], inplace=True)
display(df.head())
print(df.shape)
head = df.head(2000)
```

### Preprocessing

```python
# Quelle genre de valeurs ?
print(head[head['CGPA'].notnull()]['CGPA'].unique())
# La mediane ça donne quoi ?
print(head['CGPA'].median())
```

```python
head = df.head(2000)
# Sometimes, it needs encoding to work
# LABELISATION
le = LabelEncoder()
head['Working_Professional_or_Student_encoded'] = le.fit_transform(head['Working Professional or Student'])
head['Profession_encoded'] = le.fit_transform(head['Profession'])
head['Sleep_Duration_encoded'] = le.fit_transform(head['Sleep Duration'])
head['Dietary_Habits_encoded'] = le.fit_transform(head['Dietary Habits'])
head['Degree_encoded'] = le.fit_transform(head['Degree'])
head['City_encoded'] = le.fit_transform(head['City'])
head['Gender_encoded'] = le.fit_transform(head['Gender'])
head['Family_history_encoded'] = le.fit_transform(head['Family History of Mental Illness'])
head['Depression_encoded'] = le.fit_transform(head['Depression'])
head['suicidal_thoughts_encoded'] = le.fit_transform(head['suicidal_thoughts'])
copy = head.copy()
head = head.drop(columns=['Name','suicidal_thoughts', 'Depression','Working Professional or Student', 'Profession', 'Sleep Duration', 'Dietary Habits', 'Degree', 'City', 'Gender', 'Family History of Mental Illness'])
# NULL VALUES
head['Academic Pressure'] = head['Academic Pressure'].fillna(head['Academic Pressure'].median())
head['CGPA'] = head['CGPA'].fillna(head['CGPA'].median())
head['Study Satisfaction'] = head['Study Satisfaction'].fillna(head['Study Satisfaction'].median())
head['Job Satisfaction'] = head['Job Satisfaction'].fillna(head['Job Satisfaction'].median())
head['Work Pressure'] = head['Work Pressure'].fillna(head['Work Pressure'].median())
print(head.iloc[1])
```

### Barplot

```python
fig = plt.figure(figsize=(15,5))

# 1 ligne, 3 colonnes, emplacement 1
# WORK PRESSURE
# By GENDER
ax1 = fig.add_subplot(1, 3, 1)
plt.title("Work pressure by Gender")
x = head["Work Pressure"]
ax1.set_xlabel("Work Pressure")
ax1.set_ylabel("Index")
hue = head["Gender_encoded"]
sns.barplot(y=head.index, x=x, hue=hue)
# By DEPRESSION
ax2 = fig.add_subplot(1, 3, 2)
plt.title("Work pressure by depression")
x = head["Work Pressure"]
hue = head["Depression_encoded"]
sns.barplot(y=head.index, x=x, hue=hue)
# By SUICIDAL THOUGHT
ax3 = fig.add_subplot(1, 3, 3)
plt.title("Work pressure by Suicidal Thought")
x = head["Work Pressure"]
hue = head["suicidal_thoughts_encoded"]
sns.barplot(y=head.index, x=x, hue=hue)
```

### Heatmap

```python
numeric_columns = head.select_dtypes(include=['number'])
print(numeric_columns.index)
corr_matrix = numeric_columns.corr()
display(corr_matrix)
plt.figure(figsize=(15,6))
plt.title("Correlation between numeric columns")
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

### Scatter Plots

```python
# PLOTS
fig = plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,3,1)
plt.title("Profession per age with sleep duration")
sns.scatterplot(y=head['Profession_encoded'], x=head['Age'], hue=head['Sleep_Duration_encoded'])
ax2 = plt.subplot(1,3,2)
plt.title("Profession per age with financial stress")
sns.scatterplot(y=head['Profession_encoded'], x=head['Age'], hue=head['Financial Stress'])
ax3 = plt.subplot(1,3,3)
plt.title("Profession per age with diary habits")
sns.scatterplot(y=head['Profession_encoded'], x=head['Age'], hue=head['Dietary_Habits_encoded'])
```

### Regplots

```python
fig1 = plt.figure(figsize=(15,5))
ax1 = plt.subplot(1,2,1)
plt.title("Profession per age")
sns.regplot(y=head['Profession_encoded'], x=head['Age'])
ax2 = plt.subplot(1,2,2)
plt.title("Profession per City_encoded")
sns.regplot(y=head['Profession_encoded'], x=head['City_encoded'])
```

### Lmplot

```python
sns.lmplot(x='Age', y='Profession_encoded', hue='Depression_encoded', data=head)
plt.title("Profession per age and suicidal thoughts")
plt.show()
```

### Swarmplot

```python
sns.swarmplot(x='Depression_encoded', y='Age', hue='Gender_encoded', data=head)
plt.title('Depression_encoded by Age and Gender')
plt.show()
```
